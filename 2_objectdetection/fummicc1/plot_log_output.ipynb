```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("log_output.csv")
df = df[["epoch", "train_loss", "val_loss"]]
df["val_loss"] = df[df["val_loss"] != 0]["val_loss"]
plt.figure(facecolor="white")
plt.plot(df["epoch"], df["train_loss"], marker="o", label="train_loss")
plt.plot(df["epoch"], df["val_loss"], marker="x", label="val_loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.savefig("log_output.png")
plt.show()
```
